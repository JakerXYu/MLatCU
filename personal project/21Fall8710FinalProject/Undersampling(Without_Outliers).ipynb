## Scaling

```python
df['New_Amount'] = RobustScaler().fit_transform(df['Amount'].values.reshape(-1,1))
df['New_Time'] = RobustScaler().fit_transform(df['Time'].values.reshape(-1,1))

df.drop(['Time','Amount'], axis=1, inplace=True)
df
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
#Undersampling
from imblearn.under_sampling import RandomUnderSampler

under = RandomUnderSampler(random_state=42)
under.fit(X, y)
X_resampled, y_resampled = under.fit_resample(X, y)

df = pd.concat([X_resampled, y_resampled],axis=1)
```

## Outlier Removal: V2 V10 V14

```python
def outlier_removal(datacolumn):
    sorted(datacolumn)
    Q1,Q3 = np.percentile(datacolumn, [25,75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return lower, upper
```

```python
V2 = df['V2'].values
V2_lower, V2_upper = outlier_removal(df['V2'])
print('V2 lower and upper are', V2_lower, V2_upper)

outliers = [x for x in V2 if x < V2_lower or x > V2_upper]

print('Feature V2 Outliers: {}'.format(len(outliers)))

df = df.drop(df[(df['V2'] > V2_upper) | (df['V2'] < V2_lower)].index)
```

```python
V10 = df['V10'].values
V10_lower, V10_upper = outlier_removal(df['V10'])
print('V10 lower and upper are', V10_lower, V10_upper)

outliers = [x for x in V10 if x < V10_lower or x > V10_upper]

print('Feature V10 Outliers: {}'.format(len(outliers)))

df = df.drop(df[(df['V10'] > V10_upper) | (df['V10'] < V10_lower)].index)
```

```python
V14 = df['V14'].values
V14_lower, V14_upper = outlier_removal(df['V14'])
print('V14 lower and upper are', V14_lower, V14_upper)

outliers = [x for x in V14 if x < V14_lower or x > V14_upper]

print('Feature V14 Outliers: {}'.format(len(outliers)))

df = df.drop(df[(df['V14'] > V14_upper) | (df['V14'] < V14_lower)].index)
```
